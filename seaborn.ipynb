Seaborn:

Data visualizaton library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

seaborn has some built-in datasets

```python
#total bill vs tip dataset
tips= sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualize the data
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='smoker',size='size',col='time')
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

scatter plit

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)#here hue prints the differentspecies in different colours
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
#loading the titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

COUNT PLOT

```python
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

Bar chart

```python
sns.barplot(x='sex',y='survived',data=titanic,hue='class')
```

```python

#breast cancer diagnosis data

from sklearn.datasets import load_breast_cancer
result_cancer=load_breast_cancer()
result=pd.DataFrame(result_cancer.data,columns=result_cancer.feature_names)
result['diagnosis']=result_cancer.target
result.head()
```

```python
print(result_cancer)#data is in numpy array format
```

Distribution plot

```python

sns.distplot(result['diagnosis'])
```

Heat map

```python
correlation=result.corr()
```

```python
#constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,annot=True,cbar=True,fmt='.2f',cmap='Reds',square=True,annot_kws={'size':8})
```
